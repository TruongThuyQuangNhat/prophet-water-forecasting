```python
# Bước 1: Import các thư viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import joblib

# Bước 2: Đọc dữ liệu từ file CSV của bạn
try:
    df = pd.read_csv('data\\water_consumption_5_years.csv')
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file water_consumption_full.csv. Hãy đảm bảo file này nằm trong cùng thư mục.")
    exit()

# Bước 3: Khám phá dữ liệu (EDA)
print("5 dòng đầu tiên của dữ liệu:")
print(df.head())

print("\nThông tin tổng quan về dữ liệu:")
df.info()

print("Thống kê mô tả dữ liệu:")
print(df.describe())

# Kiểm tra giá trị thiếu
print("\nKiểm tra giá trị thiếu:")
print(df.isnull().sum())
# Xử lý giá trị thiếu nếu có (ví dụ: điền giá trị trung bình, nội suy)
# df['AvgTemperature'].fillna(df['AvgTemperature'].mean(), inplace=True)

# Bước 4: Chuyển đổi cột 'Date' sang định dạng datetime và đặt lại tên cho Prophet
df['Date'] = pd.to_datetime(df['Date'])
df_prophet = df.rename(columns={'Date': 'ds', 'WaterConsumption_m3': 'y'})

# Bước 5: Xử lý cột 'DayType' (biến phân loại)
# Prophet không tự xử lý biến phân loại. Chúng ta có thể chuyển nó thành one-hot encoding
# hoặc dùng nó làm extra regressor nếu có ý nghĩa cụ thể.
# Với DayType, có thể chuyển thành các biến boolean riêng lẻ hoặc dùng One-Hot Encoding.
# Đối với Prophet, việc thêm DayType dưới dạng regressors sẽ đòi hỏi DayType cho cả future dataframe.
# Cách đơn giản nhất là chuyển DayType thành số hoặc sử dụng trực tiếp các cờ Is_Holiday/Is_Weekend.

# Tạo thêm cờ 'Is_Weekend' từ 'DayType' nếu muốn (Prophet đã có weekly_seasonality nhưng đây là cách khác)
df_prophet['Is_Weekend'] = df_prophet['DayType'].apply(lambda x: 1 if x == 'Weekend' else 0)

# Bước 6: Trực quan hóa dữ liệu
plt.figure(figsize=(15, 7))
plt.plot(df_prophet['ds'], df_prophet['y'])
plt.title('Lịch sử tiêu thụ nước hàng ngày')
plt.xlabel('Ngày')
plt.ylabel('Lượng nước tiêu thụ (m3)')
plt.grid(True)
plt.show()

# Trực quan hóa mối quan hệ với các biến thời tiết
fig, axes = plt.subplots(3, 1, figsize=(15, 15))
sns.scatterplot(x='AvgTemperature', y='y', data=df_prophet, ax=axes[0])
axes[0].set_title('Tiêu thụ nước vs Nhiệt độ trung bình')
sns.scatterplot(x='AvgHumidity', y='y', data=df_prophet, ax=axes[1])
axes[1].set_title('Tiêu thụ nước vs Độ ẩm trung bình')
sns.scatterplot(x='Rainfall_mm', y='y', data=df_prophet, ax=axes[2])
axes[2].set_title('Tiêu thụ nước vs Lượng mưa')
plt.tight_layout()
plt.show()

# Trực quan hóa mối quan hệ với DayType và Is_Holiday
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.boxplot(x='DayType', y='y', data=df_prophet, ax=axes[0])
axes[0].set_title('Tiêu thụ nước theo loại ngày')
sns.boxplot(x='Is_Holiday', y='y', data=df_prophet, ax=axes[1])
axes[1].set_title('Tiêu thụ nước vào ngày thường và ngày lễ')
plt.tight_layout()
plt.show()

print("\nDữ liệu sau khi chuẩn bị cho Prophet (5 dòng đầu):")
print(df_prophet.head())
```

```python
# Bước 7: Chia tập dữ liệu Training và Test
# Chuỗi thời gian luôn chia theo thứ tự thời gian.
train_size = int(len(df_prophet) * 0.8) # 80% dữ liệu để huấn luyện
train_df = df_prophet.iloc[:train_size]
test_df = df_prophet.iloc[train_size:]

print(f"\nKích thước tập huấn luyện: {len(train_df)} ngày")
print(f"Kích thước tập kiểm tra: {len(test_df)} ngày")

# Bước 8: Khởi tạo và Huấn luyện mô hình Prophet
# Prophet mặc định có daily, weekly, yearly seasonality.
# Chúng ta sẽ thêm các cột thời tiết và Is_Holiday làm extra regressors.
model = Prophet(
    seasonality_mode='additive', # 'additive' hoặc 'multiplicative'
    daily_seasonality=False,     # Thường không cần nếu dữ liệu là hàng ngày và có đủ dữ liệu theo ngày trong tuần
    weekly_seasonality=True,    # Có ý nghĩa lớn
    yearly_seasonality=True     # Có ý nghĩa lớn
)

# Thêm các biến độc lập (regressors)
# Rất quan trọng: Tên cột phải khớp với tên trong DataFrame đầu vào
model.add_regressor('AvgTemperature')
model.add_regressor('AvgHumidity')
model.add_regressor('Rainfall_mm')
model.add_regressor('Is_Holiday')
# Có thể thêm 'Is_Weekend' nếu muốn tách biệt tác động của cuối tuần và ngày lễ
model.add_regressor('Is_Weekend')


# Huấn luyện mô hình
print("\nĐang huấn luyện mô hình Prophet...")
model.fit(train_df)
print("Huấn luyện mô hình hoàn tất.")

# Bước 9: Tạo DataFrame chứa các ngày cần dự đoán (bao gồm cả các ngày trong tập test)
# Prophet cần một DataFrame với cột 'ds' và các cột regressors tương ứng.
future = model.make_future_dataframe(periods=len(test_df), include_history=True)

# THÔNG TIN QUAN TRỌNG: CÁCH XỬ LÝ REGRESSORS CHO TƯƠNG LAI
# Để dự đoán, bạn PHẢI cung cấp giá trị cho 'AvgTemperature', 'AvgHumidity', 'Rainfall_mm', 'Is_Holiday', 'Is_Weekend'
# cho TẤT CẢ các ngày trong 'future' DataFrame, bao gồm cả các ngày trong tập test và các ngày tương lai bạn muốn dự đoán.
# Trong ngữ cảnh dự án này, chúng ta sẽ lấy các giá trị này từ 'df_prophet' cho các ngày đã biết.
# Nếu bạn muốn dự đoán cho những ngày HOÀN TOÀN MỚI trong tương lai, bạn sẽ cần:
#   1. Một mô hình dự báo thời tiết để dự đoán AvgTemperature, AvgHumidity, Rainfall_mm.
#   2. Một lịch ngày lễ chính xác để xác định Is_Holiday và Is_Weekend.

# Merge các cột regressors vào future DataFrame.
# Đối với các ngày trong tập huấn luyện và test, chúng ta lấy giá trị từ df_prophet.
# Đối với các ngày hoàn toàn mới (nếu bạn mở rộng future ra ngoài test_df),
# các giá trị này sẽ là NaN và bạn CẦN điền chúng (ví dụ: dự đoán, giá trị trung bình, v.v.).
# Ở đây, vì `periods=len(test_df)` và `include_history=True`, `future` sẽ khớp với `df_prophet` về các ngày,
# nên chúng ta chỉ cần merge trực tiếp.
future = pd.merge(future, df_prophet[['ds', 'AvgTemperature', 'AvgHumidity', 'Rainfall_mm', 'Is_Holiday', 'Is_Weekend']], on='ds', how='left')

# Đảm bảo không có NaN trong các cột regressor sau khi merge (chỉ kiểm tra nếu bạn không dự định mở rộng ra ngoài dữ liệu ban đầu)
# Nếu bạn mở rộng future ra ngoài dữ liệu ban đầu, bạn sẽ thấy NaN ở đây và cần xử lý.
# print("\nKiểm tra NaN trong future DataFrame sau merge (chỉ cho regressors):")
# print(future[['AvgTemperature', 'AvgHumidity', 'Rainfall_mm', 'Is_Holiday', 'Is_Weekend']].isnull().sum())

# Bước 10: Dự đoán
forecast = model.predict(future)

# Bước 11: Đánh giá mô hình trên tập kiểm tra
predictions = forecast[forecast['ds'].isin(test_df['ds'])]
actual_values = test_df['y']
predicted_values = predictions['yhat']

mae = mean_absolute_error(actual_values, predicted_values)
rmse = np.sqrt(mean_squared_error(actual_values, predicted_values))
mape = np.mean(np.abs((actual_values - predicted_values) / actual_values)) * 100

print(f"\nĐánh giá trên tập kiểm tra:")
print(f"MAE (Mean Absolute Error): {mae:.2f} m3")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f} m3")
print(f"MAPE (Mean Absolute Percentage Error): {mape:.2f}%")

# Bước 12: Trực quan hóa kết quả dự đoán
fig = model.plot(forecast)
plt.title('Dự báo tiêu thụ nước với Prophet')
plt.xlabel('Ngày')
plt.ylabel('Lượng nước tiêu thụ (m3)')
plt.show()

# Trực quan hóa các thành phần của mô hình (xu hướng, mùa vụ, tác động của regressors)
fig2 = model.plot_components(forecast)
plt.show()
```
